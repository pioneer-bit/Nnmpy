#构造随机数

```python
random_array =np.random.rand(10)
random_array
```

```python
random_array =np.random.randint(-1,10)#左边最小值，右边最大值
random_array
```

```python
random_array =np.random.randint(1,10,size=(3,3))  #构造矩阵
random_array
```

```python
np.sum(random_array) #所有数组求和
```

```python
np.sum(random_array,axis=0)    #按照第一个维度进行求和，按行求和，把每一列加在一起
```

```python
np.sum(random_array,axis=1)    #按照第一个维度进行求和，按列求和，把每一行加在一起
```

```python
random_array.ndim  #求维度
```

#矩阵的连接

```python
array1 =np.random.randint(10,size=(2,5))
array1
```

```python
array2 =np.random.randint(10,size=(2,5))
array2
```

```python
array3 = np.concatenate((array1,array2))#默认按行进行拼接
array3
```

```python
array3 = np.concatenate((array1,array2),axis=0)#默认按行进行拼接,也就是这个按第一个维度进行拼接
array3
```

```python
array3 = np.concatenate((array1,array2),axis=1)#按第二个维度进行拼接
array3
```
